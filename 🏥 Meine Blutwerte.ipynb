```python
import matplotlib.pyplot as plt
   
x = ['May' , ' June', ' July']
y1 = [5.6 ,5.31, 4.39]

normwert_1a = [4.21, 4.21, 4.21]
normwert_1b = [9.07, 9.07, 9.07]


plt.plot(x, y1, marker='o')
plt.plot(normwert_1a, "g--")
plt.plot(normwert_1b, "g--")


plt.title('1) Leukozyten', fontsize=18)
#plt.figure(figsize=(19, 19))
plt.grid(True)
plt.xlabel('Monate')
plt.ylabel('Wert')
plt.show()
```
