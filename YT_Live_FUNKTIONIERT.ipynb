```python
import pytchat
import time
import matplotlib.pyplot as plt
import numpy as np


#start a timer and count the time it takes to run the program
start_time = time.time()



#START VALUES
authors = []
messages = []
supporters = []
timestamps = []
money = []
laugh = []
lululu = []
mods = []



#ADD ID
chat = pytchat.create(video_id="INSERT_VIDEO_ID_HERE")


'''
#CONVERTS SECONDS TO MIN:SEC FORMAT
def format_time(seconds):
        seconds = int(seconds)
        minutes = seconds // 60
        seconds = seconds % 60
        return f"{minutes}:{seconds:02d}"
'''


#COUNTS THE NUMBER OF MESSAGES PER MINUTE 
#1:12 becomes 1 and 1:33:33 becomes 133

def get_minutes(timestamps):
    minutes_list = []
    for minute in timestamps:
        #if minute is 4 characters long
        if len(minute) == 4 or len(minute) == 5:
            minutes_list.append(minute.split(':')[0])

        elif len(minute) == 7:
            #input: 1:11:01
            #output: 111
            
            minutes_list.append(minute.split(':')[0] + minute.split(':')[1])
            
    return minutes_list






#calculate the total amount of money in euro
def total_amount(money):
    total = 0
    for i in money:
        total += float(i[1:])
    
    return round(total, 2)

    

    
    
    
    
    
    
def main():
    

    while chat.is_alive():
        for c in chat.get().sync_items():

            
            #TIMESTAMPS 0:00
            #time_elapsed = format_time(time.time() - start_time)
            timestamps.append(c.elapsedTime)
            
            


            #UNIQUE AUTHORS
            if c.author.name not in authors:
                authors.append(c.author.name)
            
            if c.author.isChatModerator == True and c.author.isChatModerator not in mods:
                mods.append(c.author.name)
                


            #FILTER MESSAGES
            if c.message.startswith('!') or c.message.startswith('Rolled'):
                pass
            elif '!redeem' in c.message:
                pass
            elif ', du hast' in c.message:
                pass
            elif 'Besucht unseren Discord-Server' in c.message:
                pass
            elif 'Cooldown.' in c.message:
                pass
            
            elif c.author.name == 'Streamlabs':
                pass
            
            #APPEND AND OUTPUT FILTERED MESSAGES 
            else:
                messages.append(c.message)
                print(f" {c.author.name} // {c.message} // {c.elapsedTime} // {c.amountString}")
                
#######################

            #EXTRACT LAUGHS
            if "haha" in c.message or ":rolling_on_the_floor_laughing:" in c.message or ":grinning_squinting_face:" in c.message or ":grinning_face:" in c.message or ":face_with_tears_of_joy:" in c.message or "lol" in c.message or "LOL" in c.message or "HAHA" in c.message or "XD" in c.message or "xD" in c.message or "lol" in c.message:
                laugh.append(c.elapsedTime)
                
                #print(laugh)        


            #Support
            if c.message == 'Welcome to Welpling!':
                supporters.append(c.author.name)
                

            if c.amountString == '':
                pass
            else:
                money.append(c.amountString)






#RESULTS

if __name__ == "__main__":
    main()

    print('###############################################')
    
    print('Wie viele user kommentieren?')
    print(len(authors))
    
    print('###############################################')
    
    print('Anzahl Nachrichten')
    print(len(messages))
    
    
    print('###############################################')
    
    print(len(mods))

    print(timestamps)
    print('###############################################')
    
    
    occurences = get_minutes(timestamps)
    plt.rcParams["figure.figsize"] = (20,10)
    plt.xlim(0,80)
    plt.hist(occurences, bins=200) 

    
    print('###############################################')
    #print(laugh)
    laugh_occurences = get_minutes(laugh)
    plt.hist(laugh_occurences, bins=200) 

    print('###############################################')

    #if money has no value
    print('Spenden')
    spenden = total_amount(money)
    print(spenden)
    
    print('Gibt es Aussrei√üer?')
    #print(money)
    
    print('###############################################')
    print('Welpling')
    print(len(supporters))
    print('Welpling preis total')
    print(len(supporters) * 2.99)
    #print(messages)
    
    print('###############################################')
    
```
