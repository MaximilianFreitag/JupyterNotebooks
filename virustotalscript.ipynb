```python
import os
import hashlib
import requests
import json

# specify the directory to scan
directory = "/Users/XYZ"

# set your API key
api_key = "YOUR_VIRUS_TOTAL_API_KEY"


# create a session with the VirusTotal API
session = requests.Session()
session.headers.update({"x-apikey": api_key})

output_file = "/Users/mnemo/Desktop/virus_scan_results.txt"

def scan_directory(path):
    for dirpath, _, filenames in os.walk(path):
        for filename in filenames:
            fullpath = os.path.join(dirpath, filename)
            # calculate the SHA-256 hash of the file
            with open(fullpath, "rb") as f:
                file_data = f.read()
                file_hash = hashlib.sha256(file_data).hexdigest()

            # query the VirusTotal API for information on the file
            params = {"resource": file_hash}
            response = session.get("https://www.virustotal.com/api/v3/files/{}".format(file_hash), params=params)

            # check the response from VirusTotal
            if response.status_code == 200:
                report = response.json()["data"]["attributes"]["last_analysis_stats"]
                if report["malicious"] > 0:
                    print("{}: {} positive detections out of {} total engines scanned".format(fullpath, report["malicious"], report["total"]))
                    # write to file
                    with open("virus_scan_results.txt", "a") as outfile:
                        outfile.write("{}: {} positive detections out of {} total engines scanned\n".format(fullpath, report["malicious"], report["total"]))
                else:
                    print("{}: no positive detections".format(fullpath))
            elif response.status_code == 404:
                print("{}: not found in VirusTotal database".format(fullpath))
            else:
                print("{}: error fetching report (HTTP status code {})".format(fullpath, response.status_code))

# scan the entire file system
scan_directory("/Users/mnemo")

```
