```python
from pytrends.request import TrendReq

#make a pytrends object to request Google Trends data
pytrends = TrendReq(hl='en-US')     
                
keywords = ["TikTok", "Instagram", "Reddit", "Twitter", "Snapchat"]
pytrends.build_payload(keywords, timeframe='today 5-y')
```

```python
#store kewords data
data = pytrends.interest_over_time()
data = data.drop('isPartial', axis=1)

data.tail()
```

```python
import matplotlib.pyplot as plt

#plot data
plt.plot(data)

#add titles
plt.suptitle('Social Media Searches on Google Trends')
plt.xlabel('years')                       
plt.ylabel('frequency')  

#add legend
plt.legend(keywords, loc='upper left')

plt.show()
#plt.savefig('best_language.png')
```

```python
#compare single popularity
focus = ['Instagram', 'TikTok'] 

plt.plot(data[focus])
plt.legend(focus)
```

```python
import pandas as pd

#extract country-level kewords search data
country_data = pytrends.interest_by_region(resolution='country', inc_low_vol=True)

#get countries with the highest searches of "Reddit"
data2 = country_data['Reddit'].nlargest(16)

#convert to dataframe
data2 = data2.to_frame()

data2
```

```python

#plot a bar chart
data2.plot(kind='bar', legend=None,figsize=(20,10))

#titles
plt.suptitle('Countries with the highest number of Reddit searches on Google')
plt.ylabel('Number of searches')
plt.xlabel('Countries')

plt.show()
```
