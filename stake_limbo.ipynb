```python
#flip a coin
import random
import math
```

```python


def flip_coin():
    
    coin = random.randint(0,1)
    if coin == 0:
        return 'heads'
    else:
        return 'tails'



ten_balance_outcomes = []


def calculate_odds(balance, bet):
    return (0.98 * bet)


def main():
    bet = 0.01
    balance = 10
    counter = 0
    wins = 0

    while balance > 0:
        #bet = int(input('How much do you want to bet? '))
        if balance == 100:
            wins += 1

        if bet > balance:
            print('This tactic didnt work')
            break

        coin = flip_coin()

        if coin == 'heads':
            balance += calculate_odds(balance, bet)
            counter += 1
            print(f'You WON. You got heads and your balance is now {balance}')
            bet = 0.01
        else:
            balance -= bet
            counter += 1
            print(f'You LOST. You got tails and your balance is now {balance}')
            #double the bet
            bet *= 2
            print(f'Your bet is now {bet}')


    print(f'You needed {counter} flips to loose all your money')
    ten_balance_outcomes.append(counter)



for i in range(10):
    main()



    
```

```python
#calculate the mean of ten_balance_outcomes
mean = sum(ten_balance_outcomes) / len(ten_balance_outcomes)
print(f'The mean of the ten_balance_outcomes is {mean}')
```

```python
import plotly.graph_objects as go

#visualize werte with plotly
#x axis 17 values
#y axis werte

def plot_werte(ten_balance_outcomes):
    import plotly.graph_objects as go
    import plotly.express as px
    import pandas as pd

    df = pd.DataFrame(ten_balance_outcomes)
    df.columns = ['Attempts to loose all your money']
    df['x'] = range(1, len(ten_balance_outcomes)+1)
    df['y'] = ten_balance_outcomes

    fig = px.line(df, x='x', y='y')
    fig.show()

plot_werte(ten_balance_outcomes)
```
