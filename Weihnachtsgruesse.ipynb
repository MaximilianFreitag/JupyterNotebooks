```python
from time import sleep
import random as random


christmas_message = ['Frohe Weihnachten', 'Merry Christmas', 'Joyeux Noël', 'Buon Natale', 'Feliz Navidad', 'Frohe Weihnachten','メリークリスマス', 'Wesołych Świąt', '圣诞快乐', 'Счастливого Рождества', 'God jul', 'Καλά Χριστούγεννα']
person_name = ['Rudi', 'Kai', 'Grissiii', 'Niklas']
emoji_list = ['🎅', '🎄', '🎁', '🎉', '🎆', '🎇', '🎋', '🎍', '🎎', '🎏', '🎐', '🎑', '😇', '👼 ', '❄️', '⛄ ']


def christmas_message_generator():
    import time as time

    random_message = random.choice(christmas_message)
    random_person = random.choice(person_name)
    random_emoji = random.choice(emoji_list)
    
    if random_person == 'Rudi' and random_message == 'メリークリスマス':
        random_person = 'Rudi San'

    return random_message + ' ' + random_person + ' ' + random_emoji
    time.sleep(2)



for i in range(200):
    print(christmas_message_generator())
    sleep(1)

```
