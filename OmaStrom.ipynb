```python
import matplotlib.pyplot as plt

x = [2018, 2019, 2020, 2021, 2022]
consumption = [256, 244, 419, 487, 487]
price = [145, 162, 250, 288, 324]


#make a dot at each year
plt.plot(x, consumption, 'r--', x, price, 'b--')
plt.xlabel('Jahre')
plt.ylabel('Verbrauch (kwH)')
plt.title('Oma Strom: Verbrauch vs. Jahrespreis')
plt.legend(['Verbrauch in kWh', 'Jahrespreis in Euro'])
plt.plot(x, consumption, 'ro', x, price, 'bo')
plt.axvline(x=2021, color='k', linestyle='--')

plt.show()

```
